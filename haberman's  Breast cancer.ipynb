## CDF

```python
#df_2 describe data whose status is "patients die within 5-years"
count,bin_edge=np.histogram(df_2['nodes'] ,density=True ,bins=10)
pdf=count/sum(count)
print('pdf:-',pdf)
print('bin_edge:-' ,bin_edge)
cdf=np.cumsum(pdf)
plt.xlabel('nodes')
plt.title('CDF of patient who die within 5-year')
plt.plot(bin_edge[1:],pdf ,color='red',label='pdf')
plt.plot(bin_edge[1:],cdf , label='cdf')
```

```python

count,bin_edge=np.histogram(df_1['nodes'] ,density=True ,bins=10)
pdf=count/sum(count)
print('pdf:-',pdf)
print('bin_edge:-' ,bin_edge)
cdf=np.cumsum(pdf)
plt.xlabel('nodes')
plt.title('CDF of patient who survived more than 5-year')
plt.plot(bin_edge[1:],pdf ,label='pdf' ,color='red')
plt.plot(bin_edge[1:],cdf)
```

```python
count,bin_edge=np.histogram(df_2['nodes'] ,density=True ,bins=10)
pdf=count/sum(count)
print(pdf)
print(bin_edge)
cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf)
plt.plot(bin_edge[1:],cdf)

count,bin_edge=np.histogram(df_1['nodes'] ,density=True ,bins=10)
pdf=count/sum(count)
print(pdf)
print(bin_edge)
cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf)
plt.plot(bin_edge[1:],cdf)
```

## Mean and standard-deviation

```python
print('Mean')
print(np.mean(df_1['nodes']))
print(np.mean(df_2['nodes']))
print('****************')
print(df_1.max())
print('***************')
print(df_2.max())
```

```python
#Adding outliers
print(np.mean(np.append(df_1['nodes'],100)))
print(np.mean(np.append(df_2['nodes'],110)))
```

->Mean can get affected by outliers.

```python
print('Standard-Deviation')
print(np.std(df_1['nodes']))
print(np.std(df_2['nodes']))
```

## Median ,Quantiles and Percentile

```python
print('MEDIAN')
print(np.median(df_1['nodes']))
print(np.median(df_2['nodes']))
print('***************************')
print('ADDING OUTLIERS')
print(np.median(np.append(df_1['nodes'],100)))
print(np.median(np.append(df_2['nodes'],110)))
```

->median is not getting affected by outliers.

```python
print('Quantiles')
print(np.percentile(df_1['nodes'],np.arange(0,100,25)))
print(np.percentile(df_2['nodes'],np.arange(0,100,25)))
```

```python
print(' 90th Percentile')
print(np.percentile(df_1['nodes'],90))
print(np.percentile(df_2['nodes'],90))
      
```

```python
print('Median absolute deviation(MAD) ')
from statsmodels import robust
print(robust.mad(df_1['nodes']))
print(robust.mad(df_2['nodes']))
```
